````python
cute_animals = {"rabbit", "guinea pig", "cat", "guppy"}


for animals in cute_animals:
    print(animals)
````

## Nested loops

 - a loop within a loop

```py
 (1,1)(1,2)(1,3)
 (2,1)(2,2)(2,3)
 (3,1)(3,2)(3,3)
```

````python
for x in range(3):
    for y in range(3):
        print(f"({x+1},{y+1})", end = "")
    print() # newline
````
