# while loop

 - "if-loop"
 - as long as a condition is true -> the statements inside the while loop will run.

```py
while condition:
    statements
    ...

program continues here
```


````python
i = 0
while i < 4:
    print(f"Hej {i=}")
    i +=1 # increases value by 1

print("Program continues here.")
````

## Example - Birds and oil
Oil leakage. Kills 1/2 of bird population each year. Year 0 there were 80000 birds. How many years does it take to have 1/10 remaining?

````python
birds = 80000
years = 0

while birds > 80000/10:
    birds /= 2
    years += 1
print(f"After {years} years there are {birds:.0f} birds alive")
````

## Example - Milk

 - 1L of milk has 1 500 000 bacterias in room temperature
 - Bacterias increases with 50% each hour in room temp
 - Milk get soured after 10 000 000 bacterias
 - How many hours does it take til milk sours (whole hours)?

````python
import matplotlib.pyplot as plt

amount, hours = 1.5e6, 0 # tuple unpacking 1.5e6 = 1.5 *10^6

amount_list = [amount]

print(amount_list)

while amount < 10000000:
    amount *= 1.5
    hours += 1
    amount_list.append(amount) # adds element into the list

print(amount_list)

print(f"{hours=} {amount=}")

# creates a range object with numbers: 0,1,2,...,hours
hours_list = range(0, hours+1)

plt.plot(hours_list, amount_list, 'o-')
plt.grid()
plt.xlabel("Amount of hours")
plt.ylabel("Amount of bacterias")
plt.title("Bacteras in milk in room temerature")
````
